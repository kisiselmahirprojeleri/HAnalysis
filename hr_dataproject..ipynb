```python
eksik_veri_orani = pd.isnull(df).mean() # Her sütunun eksik veri oranını yüzdesel olarak 
eksik_veri_orani
```

```python
doldurulacak_eksik_veriler = eksik_veri_orani[(eksik_veri_orani < 0.5) & (eksik_veri_orani > 0.00)].index
doldurulacak_eksik_veriler
```

```python
#sayısal eksik verileri dolduralım
def eksik_veri_doldurma(sütun:str):
    return df[sütun].fillna(df[sütun].mean(), inplace = True)

eksik_veri_doldurma('Age')
eksik_veri_doldurma('Salary')
eksik_veri_doldurma('Estimated_Annual_Bonus')
```

```python
# object sütunun eksik verisini doldurma
df['Department'].fillna(df['Department'].mode()[0], inplace= True)
```

Var olan sütunların değerlerini inceleyerek yeni sütunlar üretebilmektir.Burada yen sütun üretbilmek için aşağıdaki maddeler aranır.
     * Oluşturulan yeni sütunun veri setiyle ilişkili olması beklenir.
     * Oluşturulan yeni sütunun eski sütundan daha iyi analiz edilebilir bir yapıda olması beklenir.
     *Oluşturulan yeni sütunun eski sütundan daha zengin anlam çıkartması beklenir.

Eğer sayılabilen bir sayısal verim varsa bu veriyi belli aralıklar içinde yeni bir sütun olarak oluşturursak daha anlamlı bir bilgi çıkarabiliriz.

```python
df['Yaş_Grupları'] = pd.cut(df['Age'],[18,25,35,45,55,65], labels = ["18-25","26-35","36-45","46-55","56-65"])
df['Yaş_Grupları'].value_counts()
```
